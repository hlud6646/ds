# (1) Classification and Regression Tree (CART)
A classification tree predicts a discrete outcome (a class). A regression tree predicts a number. CART is an umbrella term.

### Pros:
- Maybe the simplest model to understand.
- Cost of using the model is logarithmic in the number of training samples.
- Supports multi-output problems.

### Cons:
- Prone to overfitting.

### Implementation notes:
- The default values will give you a huge tree (try the diabetes set without the restriction).


```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from sklearn.datasets import load_iris

iris = load_iris()
model = DecisionTreeClassifier(random_state=42, max_depth=3)
model.fit(iris.data, iris.target)

graph = plot_tree(model)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import plot_tree
from sklearn.datasets import load_diabetes

diabetes = load_diabetes()
model = DecisionTreeRegressor(random_state=42, max_depth=3)
model.fit(diabetes.data, diabetes.target)

graph = plot_tree(model)
```
